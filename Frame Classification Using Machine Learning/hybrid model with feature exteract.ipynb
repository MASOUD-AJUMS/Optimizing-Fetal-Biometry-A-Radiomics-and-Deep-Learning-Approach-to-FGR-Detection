```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import (VarianceThreshold, SelectKBest, mutual_info_classif, RFE)
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r'G:\methodical\exel\sono acoulastic\features_Opt&Subopt.csv'
df = pd.read_csv(file_path)
object_columns = df.select_dtypes(include=['object']).columns
df = df.drop(columns=object_columns)

# Separate features and labels
X = df.iloc[:, :-1].values # all rows, all columns except the last (features)
y = df.iloc[:, -1].values # the last column (labels)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# List of classifiers including CatBoost and LightGBM
classifiers = [
    LogisticRegression(), # 1
    KNeighborsClassifier(), # 2
    SVC(), # 3
    RandomForestClassifier(), # 4
    GradientBoostingClassifier(), # 5
    GaussianNB(), # 6
    DecisionTreeClassifier(), # 7
    MLPClassifier(), # 8
    AdaBoostClassifier(), # 9
    CatBoostClassifier(verbose=0), # 10 (CatBoost)
    LGBMClassifier() # 11 (LightGBM)
]

# List of feature selection and reduction methods
feature_selectors = {
    'VarianceThreshold': VarianceThreshold(threshold=0.1),
    'SelectKBest': SelectKBest(mutual_info_classif, k=10),
    'RFE': RFE(estimator=LogisticRegression(), n_features_to_select=10),
    'PCA': PCA(n_components=10),
    'LDA': LDA(n_components=1),
    'RandomForestImp': RandomForestClassifier(n_estimators=100),
    'L1_Regularization': LogisticRegression(penalty='l1', solver='liblinear'),
    'MutualInformation': SelectKBest(mutual_info_classif, k=10),
    'TreeBased': RandomForestClassifier(n_estimators=100),
    'CorrelationThreshold': VarianceThreshold(threshold=(.8 * (1 - .8))) # removing features with high correlation
}

# To store accuracy results for heatmap
results = pd.DataFrame(index=[clf.__class__.__name__ for clf in classifiers],
columns=feature_selectors.keys())

# Function to apply feature selection/reduction and train classifiers
for fs_name, selector in feature_selectors.items():
    if fs_name == 'RandomForestImp' or fs_name == 'TreeBased':
        selector.fit(X_train, y_train)
        feature_importances = selector.feature_importances_
        indices = np.argsort(feature_importances)[-10:] # Select top 10 features
        X_train_selected = X_train[:, indices]
        X_test_selected = X_test[:, indices]
    elif fs_name == 'L1_Regularization':
        selector.fit(X_train, y_train)
        selected_features = np.where(selector.coef_ != 0)[1]
        X_train_selected = X_train[:, selected_features]
        X_test_selected = X_test[:, selected_features]
    else:
        if hasattr(selector, 'fit_transform'):
            X_train_selected = selector.fit_transform(X_train, y_train)
            X_test_selected = selector.transform(X_test)
        else:
            selector.fit(X_train, y_train)
            X_train_selected = selector.transform(X_train)
            X_test_selected = selector.transform(X_test)

# Evaluate each classifier with selected features
    for clf in classifiers:
        clf.fit(X_train_selected, y_train)
        y_pred = clf.predict(X_test_selected)
        acc = accuracy_score(y_test, y_pred)
        results.loc[clf.__class__.__name__, fs_name] = acc

# Plot heatmap of the results
plt.figure(figsize=(12, 10))
sns.heatmap(results.astype(float), annot=True, cmap='viridis', cbar_kws={'label': 'Precision'})
# sns.heatmap(results.astype(float), annot=True, cmap='viridis', cbar_kws={'label': 'F1 Score'})
# sns.heatmap(results.astype(float), annot=True, cmap='viridis', cbar_kws={'label': 'Accuracy'})
# sns.heatmap(results.astype(float), annot=True, cmap='viridis', cbar_kws={'label': 'AUC'})
plt.title('Classification Precision for Different Feature Selection Methods')
plt.ylabel('Classifiers')
plt.xlabel('Feature Selection Methods')
plt.show()
```
